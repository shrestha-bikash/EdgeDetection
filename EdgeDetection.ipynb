# Load the data and necessary modules

```
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import cv2

url = 'https://raw.githubusercontent.com/badriadhikari/2019-Fall-AI/master/homeworks/one-channel.csv'
image2D = np.loadtxt(url, delimiter=",")

print(image2D)
```

# Custom function for doing covolution

```
def convolution2D(image2D, kernel3x3):
  # convolved2D = cv2.filter2D(image2D, -1, kernel3x3)
    m, n = kernel3x3.shape
    y, x = image2D.shape
    if (m == n):
        # padding with the edge value of the array (option 1)
        image = np.pad(image2D, 1, mode='edge')

        # padding all sides with the 0 value (option 2)
        # image = np.pad(image2D, 1)

        convolved2D = np.zeros((y,x))

        for i in range(y):
            for j in range(x):
                convolved2D[i][j] = np.sum(image[i:i+m, j:j+m]*kernel3x3)
  
    return convolved2D
```

# Generate the output

```
sns.heatmap(image2D, cmap='Greens')
plt.show()

edge_detect_filter_3x3 = np.array([[-1, -1, -1], [-1, 8, -1], [-1, -1, -1]])

# Convolve once
convolved_image = convolution2D(image2D, edge_detect_filter_3x3)

sns.heatmap(convolved_image, cmap='Greens')
plt.show()

# # Convolve again
convolved_image = convolution2D(convolved_image, edge_detect_filter_3x3)

sns.heatmap(convolved_image, cmap='Greens')
plt.show()
```
